3. Estimar el puntaje Z de una variable (2 puntos): Se debe utilizar la que posea la menor cantidad de datos perdidos 
(o que no posea).
‚óè Generar una nueva columna en la tabla de datos que contenga el puntaje Z de dicha variable. Recuerde que para calcular
el puntaje z de una observaci√≥n se utiliza la f√≥rmula: ùëßùëñ = ( ùë•ùëñ ‚àí ùë• ) / œÉ (ùë•)
‚óè Para ordenar los valores pueden utilizar la funci√≥n sort_values del DataFrame analizado. Lean la documentaci√≥n asociada 
al argumento by.
‚óè Aseg√∫rense de eliminar las observaciones que tengan valores perdidos con dropna. Lean la documentaci√≥n asociada al
argumento subset.
‚óè Seleccionen las primeras y √∫ltimas 10 observaciones de la base reordenada.
¬øQu√© caracter√≠sticas en com√∫n se encuentran?


```python
#Esta funci√≥n me permite calcular el puntaje Z  de cualquier columna
def get_zscore(col, case):
    get_point = df[df['cname']==case][col]
    preproc=df[col].dropna()
    get_mean = preproc.mean()
    get_std = np.sqrt(preproc.var())
    return (get_point - get_mean) / get_std
```

```python
#Probando para la columna escogida con dos paises especificos
get_zscore('gle_cgdpc', 'Yemen')
```

```python
get_zscore('gle_cgdpc', 'Haiti')
```

```python
#Apliqu√© el dropna aunque esta columna no tiene nulos, segun el paso anterior. Lo hice para plicar el subset

df_z = df.dropna(subset=['gle_cgdpc']) 
df_z
```

```python
#agregar√© la columna Puntaje Z
df_z = df_z.assign(puntaje_Z=0)
df_z
```

```python
#Ahora rellenar√© la columna puntajeZ con un for
preproc=df['gle_cgdpc'].dropna() #Esto es parte de la funci√≥n get_zscore pero ahora lo oupar√© para la columna gle_cgdpc en especifico
get_mean = preproc.mean()  #porque necesitar√© el get_mean y el get_std en el for
get_std = np.sqrt(preproc.var())

for i in range(len(df_z)):
    df_z.loc[i,'puntaje_Z'] = (df_z.iloc[i,5]-get_mean)/get_std
df_z
```

```python
#Ahora voy a ordenar el df seg√∫n la columna puntajeZ
df_z2= df_z.sort_values(by= ['puntaje_Z']).reset_index(drop=True)
df_z2
```

5. Generar curvas de Densidad de Probabilidad (2 puntos).
‚óè Cree una funci√≥n que permita visualizar las curvas de Densidad de Probabilidad
para las variables imf_pop, gle_cgdpc y wdi_exph.
‚óè La funci√≥n debe contener como argumentos:
‚óã df: Tabla de datos 1.
‚óã variable: Variable a contrastar.

```python
#Definimos la funci√≥n curvas de densida de probabilidad como cur_dens_prob
def cur_dens_prob(df, var):
    densidad = df[var].plot.density()
    plt.show(densidad)
```

```python
#Voy a probar la funci√≥n para gle_cgdpc
cur_dens_prob(df_z2, 'gle_cgdpc')
```

```python
#Ahora para ver con las variables: imf_pop, gle_cgdpc y wdi_expedu, voy a hacer un for
for var in ['imf_pop', 'gle_cgdpc','wdi_expedu']:
    print('Curva de densidad de Probabilidad para', var)
    cur_dens_prob(df_z2, var)
```

```python
# Generamos la normalizaci√≥n del gdp mediante el logaritmo.
gdp_dropna = df_z2['gle_cgdpc'].dropna()
log_gdp = np.log(gdp_dropna)
print(round(log_gdp.mean() ,2))
```

```python
plt.subplot(1, 2, 1)
plt.hist(gdp_dropna, color='grey', alpha=.4, density=True)
x_min, x_max = plt.xlim()
x_axis = np.linspace(x_min, x_max, 100)

plt.plot(x_axis, stats.norm.pdf(x_axis,gdp_dropna.mean(),gdp_dropna.std()),color='tomato', lw=3)
plt.axvline(gdp_dropna.mean(),color='dodgerblue',linestyle='--', lw=3)
plt.title("Histograma PIB")

plt.subplot(1, 2, 2)
plt.hist(log_gdp, color='grey', alpha=.4, density=True)
x_min, x_max = plt.xlim()
x_axis = np.linspace(x_min, x_max, 100)
plt.plot(x_axis, stats.norm.pdf(x_axis,log_gdp.mean(),log_gdp.std()),color='tomato', lw=3)
plt.axvline(log_gdp.mean(),color='dodgerblue',linestyle='--', lw=3)
plt.title("Histograma log(PIB)");
```
